```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
# Define ticker
tickerSymbol = 'NVDA'

# Get ticker data
tickerData = yf.Ticker(tickerSymbol)

# Get balance sheet
balance_sheet = tickerData.balance_sheet

# Get key stats
key_stats = tickerData.info
key_stats_df = pd.DataFrame(list(key_stats.items()),columns = ['Attribute','Value'])


# Get current price
current_price = tickerData.info['currentPrice']

# Calculate valuation metrics
pe_ratio = current_price / key_stats['trailingPE']
pb_ratio = current_price / key_stats['priceToBook']
ps_ratio = current_price / key_stats['priceToSalesTrailing12Months']

# Define threshold values
pe_avg = 25
pb_avg = 3
ps_avg = 5

# Check if under or overvalued
if pe_ratio < pe_avg:
    print("P/E ratio is below average, stock may be undervalued")
else:
    print("P/E ratio is above average, stock may be overvalued")

if pb_ratio < pb_avg:
    print("P/B ratio is below average, stock may be undervalued")
else:
    print("P/B ratio is above average, stock may be overvalued")

if ps_ratio < ps_avg:
    print("P/S ratio is below average, stock may be undervalued")
else:
    print("P/S ratio is above average, stock may be overvalued")

# Extract specific metrics
growth = key_stats['revenueGrowth']
profit_margins = key_stats['profitMargins']
valuation = key_stats['marketCap']
returns = key_stats['returnOnAssets']
debts = key_stats['totalDebt']
ebit = key_stats['ebitda']

# Get historical data
end_date = datetime.today().strftime('%Y-%m-%d')
start_date = (datetime.today() - timedelta(days=365)).strftime('%Y-%m-%d')
hist = tickerData.history(start=start_date, end=end_date)

# Get weekly prices
weekly_prices = hist['Close'].resample('W').last()

# Print results
print(balance_sheet)
print(key_stats_df)
print(weekly_prices)
```
